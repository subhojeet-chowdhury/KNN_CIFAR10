```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import os
import platform

%matplotlib inline

```

# KNN model

```python
class CustomKNN(object):
    def __init__(self):
        pass

    def fit(self, training_data, labels):
        """
        Fit the KNN model with training data and corresponding labels.

        Parameters:
        - training_data: The feature vectors of the training data.
        - labels: The labels corresponding to the training data.
        """
        self.training_data = training_data
        self.labels = labels

    def predict(self, test_data, k=1, use_optimized=True):
        """
        Predict labels for test data using KNN.

        Parameters:
        - test_data: The feature vectors of the test data.
        - k: The number of neighbors to consider.
        - use_optimized: Flag to choose between optimized and non-optimized distance calculation.

        Returns:
        - predicted_labels: Predicted labels for the test data.
        """
        if use_optimized:
            distances = self.calculate_distances_optimized(test_data)
        else:
            raise ValueError('Invalid value for optimization flag')
        return self.predict_labels(distances, k=k)

    def calculate_distances_optimized(self, test_data):
        """
        Calculate distances between test data and training data using an optimized approach.

        Parameters:
        - test_data: The feature vectors of the test data.

        Returns:
        - distances: Matrix of distances between test and training data.
        """
        num_test = test_data.shape[0]
        num_train = self.training_data.shape[0]
        distances = np.zeros((num_test, num_train))
        distances = np.sqrt(np.sum(np.square(self.training_data), axis=1) +
                            np.sum(np.square(test_data), axis=1)[:, np.newaxis] -
                            2 * np.dot(test_data, self.training_data.T))
        return distances

    def predict_labels(self, distances, k=1):
        """
        Predict labels based on distances and the k-nearest neighbors.

        Parameters:
        - distances: Matrix of distances between test and training data.
        - k: The number of neighbors to consider for prediction.

        Returns:
        - predicted_labels: Predicted labels for the test data.
        """
        num_test = distances.shape[0]
        predicted_labels = np.zeros(num_test)
        for i in range(num_test):
            closest_labels = []
            sorted_distances = np.argsort(distances[i])
            closest_labels = list(self.labels[sorted_distances[0:k]])
            predicted_labels[i] = np.argmax(np.bincount(closest_labels))
        return predicted_labels
```

# Helper functions for loading data

```python
def load_pickled_data(file_obj):
    """
    Load pickled data from a file, adjusting for Python version.

    Parameters:
    - file_obj: File object of the pickle file.

    Returns:
    - Unpickled data.
    """
    python_version = platform.python_version_tuple()
    if python_version[0] == '2':
        return pickle.load(file_obj)
    elif python_version[0] == '3':
        return pickle.load(file_obj, encoding='latin1')
    raise ValueError("Unsupported Python version: {}".format(python_version))

def load_cifar_batch(file_path):
    """
    Load a single batch of CIFAR data from a file.

    Parameters:
    - file_path: Path to the CIFAR batch file.

    Returns:
    - X: Array of data, reshaped and transposed.
    - Y: Array of labels.
    """
    with open(file_path, 'rb') as file_obj:
        batch_data = load_pickled_data(file_obj)
        X_data = batch_data['data']
        Y_labels = batch_data['labels']
        X_data = X_data.reshape(10000, 3, 32, 32).transpose(0, 2, 3, 1).astype("float")
        Y_labels = np.array(Y_labels)
    return X_data, Y_labels

def load_cifar10_data(root_directory):
    """
    Load all batches of CIFAR data.

    Parameters:
    - root_directory: Root directory of CIFAR data.

    Returns:
    - X_train: Training data.
    - Y_train: Training labels.
    - X_test: Test data.
    - Y_test: Test labels.
    """
    train_batches = []
    train_labels = []

    # Load training batches
    for batch_number in range(1, 6):
        file_path = os.path.join(root_directory, 'data_batch_%d' % (batch_number,))
        X_train_batch, Y_train_batch = load_cifar_batch(file_path)
        train_batches.append(X_train_batch)
        train_labels.append(Y_train_batch)

    # Concatenate training data and labels
    X_train = np.concatenate(train_batches)
    Y_train = np.concatenate(train_labels)

    # Load test batch
    X_test, Y_test = load_cifar_batch(os.path.join(root_directory, 'test_batch'))

    return X_train, Y_train, X_test, Y_test
```

# Loading raw files and reading them as training and testing datasets

```python
# Specify the directory containing CIFAR-10 data
cifar10_directory = './data/cifar-10-batches-py'

# Load CIFAR-10 data using the previously defined function
X_train_data, y_train_labels, X_test_data, y_test_labels = load_cifar10_data(cifar10_directory)

# Display the size of the training and testing datasets
print('Training data shape: ', X_train_data.shape)
print('Training labels shape: ', y_train_labels.shape)
print('Test data shape: ', X_test_data.shape)
print('Test labels shape: ', y_test_labels.shape)
```

# Visualizing dataset samples

```python
# Define the CIFAR-10 classes
classes = ['plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

# Get the number of classes
num_classes = len(classes)

# Specify the number of samples to display per class
samples_per_class = 7

# Loop over each class
for y, cls in enumerate(classes):
    # Get the indices of examples belonging to the current class
    idxs = np.flatnonzero(y_train_labels == y)
    
    # Randomly select samples_per_class examples from the current class
    idxs = np.random.choice(idxs, samples_per_class, replace=False)
    
    # Loop over selected indices
    for i, idx in enumerate(idxs):
        # Calculate the subplot index
        plt_idx = i * num_classes + y + 1
        
        # Create a subplot
        plt.subplot(samples_per_class, num_classes, plt_idx)
        
        # Display the image
        plt.imshow(X_train_data[idx].astype('uint8'))
        
        # Turn off axis labels
        plt.axis('off')
        
        # Add class name as title for the first subplot in each class
        if i == 0:
            plt.title(cls)

# Show the plot with the sampled images
plt.show()
```

# Data subsampling

```python
# Set the desired number of training samples
num_training_samples = 10000

# Create a mask with indices up to num_training_samples
training_mask = list(range(num_training_samples))

# Subsample the training data and labels using the mask
X_train_subsampled = X_train_data[training_mask]
y_train_subsampled = y_train_labels[training_mask]

# Set the desired number of test samples
num_test_samples = 1000

# Create a mask with indices up to num_test_samples
test_mask = list(range(num_test_samples))

# Subsample the test data and labels using the mask
X_test_subsampled = X_test_data[test_mask]
y_test_subsampled = y_test_labels[test_mask]
```

```python
# Reshape the training data to have each sample as a row
X_train_reshaped = np.reshape(X_train_subsampled, (X_train_subsampled.shape[0], -1))

# Reshape the test data to have each sample as a row
X_test_reshaped = np.reshape(X_test_subsampled, (X_test_subsampled.shape[0], -1))

# Display the shapes of the reshaped training and test data
print('Reshaped Training data shape: ', X_train_reshaped.shape)
print('Reshaped Test data shape: ', X_test_reshaped.shape)
```

# Performing K-NN

```python
# Create an instance of the CustomKNN classifier
custom_knn_classifier = CustomKNN()

# Train the classifier with the reshaped training data
custom_knn_classifier.fit(X_train_reshaped, y_train_subsampled)

# Compute distances between test data and training data
distances = custom_knn_classifier.calculate_distances_optimized(X_test_reshaped)

# Predict labels for the test data using the trained classifier
y_test_predicted = custom_knn_classifier.predict_labels(distances, k=5)

# Calculate the number of correctly predicted labels
num_correct_predictions = np.sum(y_test_predicted == y_test_subsampled)

# Calculate the accuracy
accuracy = float(num_correct_predictions) / num_test_samples

# Display the results
print('Got %d / %d correct => accuracy: %f' % (num_correct_predictions, num_test_samples, accuracy))
```

# Cross validation

performing 5-fold cross validation on k-NN for varying values of k 

```python
# Number of folds for cross-validation
num_folds = 5

# List of different k values to evaluate
k_choices = [1, 3, 5, 8, 10, 12, 15, 20, 50, 100]

# Split the reshaped training data and labels into 'num_folds' folds
X_train_folds = np.array_split(X_train_reshaped, num_folds)
y_train_folds = np.array_split(y_train_subsampled, num_folds)

# Dictionary to store accuracies for different k values
k_to_accuracies = {}

# Loop over each k value
for k in k_choices:
    k_to_accuracies[k] = []
    
    # Loop over each fold
    for num_knn in range(0, num_folds):
        # Separate current fold as test data
        X_test_fold = X_train_folds[num_knn]
        y_test_fold = y_train_folds[num_knn]
        
        # Combine remaining folds as training data
        X_train_fold = np.delete(X_train_folds, num_knn, axis=0)
        X_train_fold = np.concatenate(X_train_fold, axis=0)
        
        y_train_fold = np.delete(y_train_folds, num_knn, axis=0)
        y_train_fold = np.concatenate(y_train_fold, axis=0)
        
        # Create and train the CustomKNN classifier
        classifier = CustomKNN()
        classifier.fit(X_train_fold, y_train_fold)
        
        # Compute distances and predict labels for the test data
        dists = classifier.calculate_distances_optimized(X_test_fold)
        y_test_pred = classifier.predict_labels(dists, k)

        # Calculate accuracy and store it in the dictionary
        num_correct = np.sum(y_test_pred == y_test_fold)
        accuracy = float(num_correct) / num_test_samples
        k_to_accuracies[k].append(accuracy)

# Display the accuracies for different k values and folds
print("Printing our 5-fold accuracies for varying values of k:")
print()
for k in sorted(k_to_accuracies):
    for accuracy in k_to_accuracies[k]:
        print('k = %d, accuracy = %f' % (k, accuracy))
```

# Displaying cross-validation results for various k values

```python
# Plotting individual data points for each k with accuracies
for k_val in k_choices:
    accuracy_values = k_to_accuracies[k_val]
    plt.scatter([k_val] * len(accuracy_values), accuracy_values, label=f'k={k_val}', alpha=0.7)

# Calculating mean and standard deviation of accuracies for each k
mean_accuracies = np.array([np.mean(val) for k, val in sorted(k_to_accuracies.items())])
std_accuracies = np.array([np.std(val) for k, val in sorted(k_to_accuracies.items())])

# Plotting a trend line with error bars corresponding to standard deviation
plt.errorbar(k_choices, mean_accuracies, yerr=std_accuracies, linestyle='-', marker='o', markersize=8, capsize=5, color='black')

# Setting plot labels and title
plt.title('Cross-validation Performance for Different k Values')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.legend(loc='upper right')  # Adding legend for individual k values

# Enhancing the plot 
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

# Adding additional information or insights about the plot
plt.annotate('Trend Line with Error Bars', xy=(10, 0.85), xytext=(30, 0.82),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Show the plot with improved aesthetics
plt.show()
```

# Choosing the best value of k based on cross-validation results

```python

best_k = 20

# Creating an instance of the CustomKNN classifier
classifier = CustomKNN()

# Training the classifier with the reshaped and subsampled training data
classifier.fit(X_train_reshaped, y_train_subsampled)

# Predicting labels for the reshaped test data using the chosen k value
y_test_predictions = classifier.predict(X_test_reshaped, k=best_k)

# Computing and displaying the accuracy for the chosen k value on the test set
num_correct = np.sum(y_test_predictions == y_test_subsampled)
accuracy = float(num_correct) / num_test_samples
print(f'Obtained {num_correct} / {num_test_samples} correct => Accuracy: {accuracy:.4f}')
```
